```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
def fn(x):
    return x**2*np.sin(x)
y=fn(x)
plt.plot(x,y)
```

```python
#functions and their derivatives
X=np.linspace(1,10,1000)
Y=(lambda x : x**2)(X)
dydx=np.gradient(Y,X)
y_int=np.cumsum(Y)*abs(X[1]-X[0])
plt.plot(X,Y,label="fn")
plt.plot(X,y_int,label="integral")
plt.plot(X,dydx,label="dydx")
plt.legend()
```

```python
X=np.linspace(-10,10,1000)
Y=(lambda x:x**3)(X)
dydx=np.gradient(Y,X)
plt.plot(X,Y,label="fn")
plt.plot(X,dydx,label="dydx")
plt.legend()
```

```python
X=np.linspace(1,10,10000)
Y=np.exp(-X/10)*np.sin(X)
dydx=np.gradient(Y,X)
plt.plot(X,Y,label="fn")
plt.plot(X,dydx,label="dydx")
plt.legend()
```

```python

#polar co-ordinates
X=np.linspace(1,10,1000)
x=3+4*np.sin(X)
y=np.cos(X)
plt.plot(x,y)
```

```python
#multivariate fns
x=np.linspace(-10,10,1000)
y=x
xv,yv=np.meshgrid(x,y)
zv=xv**2+yv**2
plt.contourf(xv,yv,zv,levels=20)
```

```python
x=np.linspace(-2,2,1000)
y=x
xv,yv=np.meshgrid(x,y)
zv=np.exp(-(xv**2+yv**2))*np.sin(xv)
plt.contourf(xv,yv,zv,20)
plt.colorbar()
```
