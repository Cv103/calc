```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.axes(projection="polar")
r=2
rads=np.arange(0,2*(np.pi),0.01)
r=1+np.cos(rads)
plt.polar(rads,r,)
```

```python
plt.axes(projection="polar")
a=2
rads=np.arange(0,2*np.pi,0.01)
r=a+a*np.sin(rads)
plt.plot(rads,r)
```

```python
plt.axes(projection="polar")
rads=np.arange(0,8*np.pi,0.01)
r=np.sin(2.4*rads)**2+np.cos(2.4*rads)**4
plt.polar(rads,r)
```

```python
plt.axes(projection="polar")
rads=np.arange(0,12*np.pi,0.05)
r=np.sin(1.2*rads)**2+np.cos(6*rads)**3
plt.polar(rads,r)
```

```python
t=np.arange(0,12*np.pi,0.02)
x=np.sin(t)-np.sin(2.3*t)
y=np.cos(t)
plt.plot(x,y)
```

```python
t=np.arange(-2,2,0.01)
x=t**2
y=t**3-3*t
rng=np.linspace(2,4,100)
eq=(3**0.5)*(rng-3)
plt.plot(rng,eq,'r')
plt.plot(x,y)
```

```python
t=np.arange(-2,4,0.01)
x=t**2-2*t
y=t+1
rng=np.linspace(-2,2,100)
eq=-(rng/2 )+1
plt.plot(x,y)
plt.plot(rng,eq,label="tangent")
r1=np.linspace(0,6,100)
eq1=(1/4)*r1+(13/4)
plt.plot(r1,eq1,label="tangent")
plt.legend()
```

```python
t=np.arange(0,2*np.pi,0.01)
x=np.sin(2*t)
y=np.cos(2*t)
rng=np.linspace(-.75,.75,100)
eq=(rng*0)+1
plt.plot(rng,eq)
plt.plot(x,y)
```
